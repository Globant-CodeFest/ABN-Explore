```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 200)

# Define column names
col_names = ['ID', "LATITUDE", "LONGITUDE", "ELEVATION", "STATION"]
# Define value and flag names for each month

# Define column widths
col_widths = [12, 8, 10, 8, 25]

# Read the file
df = pd.read_fwf('DATA/ghcnm.tavg.v4.0.1.20230609.qfe.inv', widths=col_widths, names=col_names)
```

```python
df.head()
```

```python
# Replace first column with the first 2 characters of that same column
df['ID'] = df['ID'].str[:2]
```

```python
df.head()
```

```python
countryCodes = pd.read_csv('COUNTRIESDATA/COUNTRIES.csv')
countryCodes.head()
```

```python
# Merge the dataframes and drop the ID column
df = pd.merge(df, countryCodes, left_on='ID', right_on='Country Code').drop('ID', axis=1)
```
